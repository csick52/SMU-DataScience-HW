WeatherPy

```python
# Import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
import numpy as np
import json

# Import citipy
from citipy import citipy

# Import API key
from api_keys import api_key
```

Generate a List of Cities

```python
# Make 2 lists
# One list will hold randomly generated coordinates
# One list will hold the cities found using those coordinates 
coords = []
cities = []

# Generate the random coordinates
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
coords = zip(lats, lngs)

# Use citipy and the coordinates to fill the list with cities
for loc in coords:
    city = citipy.nearest_city(loc[0], loc[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

Perform API Calls

```python
# Create lists to hold relevant weather data
temperature = []
humidity = []
cloudiness = []
wind_speed = []
latitude = []

# OpenWeatherMapAPI doesn't have a record of some cities
# Make an list to store the cities whose data we do collect
cities_used = []

# Iterate through cities list, updating the query url as we go
count = 0
for the_city in cities:
    query_url = (f"http://api.openweathermap.org/data/2.5/weather?q={the_city}&APPID={api_key}")
    response = requests.get(query_url).json()
    
    # Use try/except to avoid KeyErrors where OpenWeatherMapAPI has no records
    try:
        response["id"]
        
    except KeyError:
        print("City not found. Skipping...")
        continue
        
    # Save the weather data to the lists above
    temperature.append(response["main"]["temp"])
    humidity.append(response["main"]["humidity"])
    cloudiness.append(response["clouds"]["all"])
    wind_speed.append(response["wind"]["speed"])
    latitude.append(response["coord"]["lat"])
    cities_used.append(the_city.title())
    
    # Increase the count
    count += 1
        
    # Print log of each city in the list
    print(f"Processing record {count} | " + (the_city).title())
    
    # Break the loop once data on 500 cities is collected
    if count == 500:
        break
```

Convert Raw Data to Dataframe

```python
# Change the temperature from Kelvin to Farenheit: (temp − 273.15) × 9/5 + 32
temps_f = [round(((i - 273.15) * (9/5)) + 32, 2) for i in temperature]

# Change wind speed from meters per second to miles per hour
wind_speed_mph = [round((i * 2.237), 2) for i in wind_speed]

# Convert the data into a dictionary
cityWeather_dict = {
    "City":cities_used,
    "Latitude":latitude,
    "Temperature (F)":temps_f,
    "% Humidity":humidity,
    "% Cloudiness":cloudiness,
    "Wind Speed (MPH)":wind_speed_mph
}

# Convert the dictionary to a dataframe using pandas
cityWeather = pd.DataFrame(cityWeather_dict)
cityWeather.head()
cityWeather.to_csv("cityWeather_df.csv", sep='\t')
```

Plot the Data

```python
# Temperature vs Latitude
x = cityWeather["Latitude"].tolist()
y = cityWeather["Temperature (F)"]
temp_lat_plot = plt.scatter(x=x, y=y, c="skyblue", edgecolors="black")
plt.title("Temperature vs Latitude 1/11/2019")
plt.xlabel("Latitude")
plt.ylabel("Temp. (F)")
plt.grid()
plt.show()
plt.savefig('temperature.png', bbox_inches='tight')
```
